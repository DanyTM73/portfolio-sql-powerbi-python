# THE *WHILE* SENTENCE

*While* is a control statement that allows executing a portion of code repeatedly **as long as** a certain condition is met.

## Structure

```python
while condition:
    # lines of code to execute
```

As long as the `condition` has a `True` value, the `#lines of code to be executed` will run. Since the code is executed repeatedly, the *while* statement is known as a **cycle** or ***loop***.

## Examples

```python
# The simplest example of all: setting 'condition' to 'True', which will cause the code to run an infinite number of times.
while True:
    print('Running code')
    print('Infinite loop')
```

```python
# The opposite example to the previous one: a while loop that NEVER executes because the condition is always false.
while False:
    print('Running code: the loop never runs')
```

```python
# A more elaborate example: a savings account with an initial balance and each day we make a withdrawal.
balance = 1000
day = 0
while balance >= 0:
    # Update the day
    day = day + 1 # Hablar de dia += 1
    
    # Input: the amount to withdraw
    withdraw = int(input('Input the amount to withdraw: '))
    
    # Update balance
    balance = balance - withdraw
    
    # Print day and available balance
    print(f'Day: {day}. Available balance: {balance}')
    
```

Let us observe that depending on the amounts withdrawn, the number of days (that is, the number of repetitions of the code) will change. Therefore, *while* is a loop where we generally do not know in advance how many times it will repeat.

```python
# We can even include "if" statements in combination with "while" statements. 
# Example: if the balance is 0 or negative, let's print a message on the screen.
balance = 1000
day = 0
while balance >= 0:
    # Update day
    day = day + 1 # Talk about day += 1
    
    # Input: the amount to withdraw
    withdraw = int(input('Enter the amount to withdraw: '))
    
    # Update balance
    balance = balance - withdraw
    
    # Print day and available balance
    print(f'Day: {day}. Available balance: {balance}')
    
    # Print message if balance is below or equal to 0
    if balance <= 0:
        print('ALERT: The balance is 0 or negative!')
```
